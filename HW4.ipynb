```python
def min_edit_dist(str1, str2, m, n):
    if m == 0:
        return n
    if n == 0:
        return m
    if str1[m - 1] == str2[n - 1]:
        return min_edit_dist(str1, str2, m-1, n-1)
    
    return 1 + min(min_edit_dist(str1, str2, m-1, n-1),
                   min_edit_dist(str1, str2, m-1, n),
                   min_edit_dist(str1, str2, m, n-1))
```
